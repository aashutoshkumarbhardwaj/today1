```python
import numpy as np
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/machine learning/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('passenger travelling')
```

```python
#extract numerocal part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
#extract categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_numerical']=df['Cabin'].str.extract('(\d+)')
df['cabin_categorical']=df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_categorical'].value_counts().plot(kind='bar')
```

```python
df['cabin_categorical'].value_counts()
```

```python
df['ticket_numerical']=df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_numerical']=pd.to_numeric(df['ticket_numerical'],errors='coerce',downcast='integer')
```

```python
df['ticket_categorical']=df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_categorical']=np.where(df['ticket_numerical'].isnull(),df['ticket_categorical'],np.nan)
```
